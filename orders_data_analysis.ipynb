```python
import pandas as pd
df = pd.read_csv("data/orders.csv")
df.head(10)
```

```python
df = pd.read_csv("data/orders.csv",na_values=['Not Available','unknown'])
df['Ship Mode'].unique()  
```

```python
df.head(10)
```

```python
#lowercase and replace space with underscore
df.columns=df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.head(5)
```

```python
#derive new columns discount, sale_price, profit
df['discount'] = df['list_price'] * df['discount_percent'] / 100
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df.head(4)
```

```python
#drop cost_price, list_price and discount_percent columns
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
df
```

```python
#convert order date from object(str) data type to datetime data type
df.dtypes
df['order_date'] = pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```
